```python
import os
import json
directory = r"C:/Users/james\Desktop/Train_steering_wheel/make_yolo/ann"
final_dict = {}
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    if os.path.isfile(f):
        with open(f) as jsonfile:
            ann_dict = json.load(jsonfile)
            if ann_dict["objects"] != []:
                final_dict[filename] = ann_dict["objects"][0]["points"]
                final_dict[filename]["size"] = ann_dict["size"]

final_dict
```

Create dict with list of normalized coordinates to form [x1, y1, x2, y2]
where 1 denotes top left and 2 denotes bottom right

```python
create_norm_dict = {}

for keys in final_dict.keys():
    img = keys[:len(keys)-9]
    bb = final_dict[keys]["exterior"]
    h_w = list(final_dict[keys]["size"].values())
    ux = bb[0][0]/h_w[1] # top left x
    uy = bb[0][1]/h_w[0] # top left y
    lx = bb[1][0]/h_w[1] # bottom right x 
    ly = bb[1][1]/h_w[0]    # bottom right y

    create_norm_dict[img] = [ux,uy,lx,ly]
create_norm_dict
```

Convert to [object, center point, height, width]
in form [0, x, y, w, h]

```python
xywh_dict = {}
for keys in create_norm_dict.keys():
    coord_list = create_norm_dict[keys]
    w = coord_list[2]-coord_list[0]
    h = coord_list[3]-coord_list[1]
    x = (coord_list[0]+coord_list[2])/2
    y = (coord_list[1]+coord_list[3])/2
    xywh_dict[keys] = [0, x, y, w, h]
```

Create txt files


```python

for keys in xywh_dict.keys():
    txt_filename = keys + ".txt"
    filename = "C:/Users/james\Desktop/Train_steering_wheel/make_yolo/txtfiles/" + txt_filename
    file = open(filename,'w')
    for item in xywh_dict[keys]:
        file.write(str(item) + " ")
    file.close()
```
